# 57) 2025-11-01 17:56, POC алфавита фичей

**Идея**: получить компактный и ограниченный фича спейс на базе кондишнинга. По-другому - получить алфавит (репертуар) фичей.

Делаем:
1) заводим словарь:
    - ключ: OHE вектор, для обозначения кода фичи
    - значение: "гармошка" или real-valued вектор, где каждое real-value показывает удалённость, например, горизонтальной палки от центра
2) фаза бордствования:
    - предъявляем изображения цифр
    - по каждой из цифре вычисляем "гармошку" - это типа фича или cue.
    - запоминаем фичу под новым разрядом OHE, также запоминаем связку с цифров
    - когда весь OHE вектор забился - остановка 
4) фаза сна:
    - с помощью reverse search находим коррелированние фичи
    - объединяем коррелированные фичи, смотрим, чтобы не нарушалась связка с цифрой. В случае нарушения связки откатываем объединение. Пре успешном объединении вместо двух фичей, остаётся одна
    - в конце фазы переиндексируем OHE фичи
   
Ожидается, что через череду фаз "бодрствование" - "сон" мы выйдем на компактный набор фичей, которые достаточно для дискрминации циферок.

Про формулу Байеса и применимость её в задаче оценки параметров модели  
P(θ|x) = <font color="green">P(x|θ)</font> * <font color="red">P(θ)</font> / <font color="purple">P(x)</font>

Итак. Представим, что у нас один параметр θ и одна переменная Х (скаляр).
Тогда можно составить JPT, как показано на рисунке слева. При этом:
- сразу понимаем, что это виртуальная таблица, т.к. физически её очень сложо создать из-за размерности - экспоненциальный взрыв
- пример нарисован для дискретного случая. Для непрерываного суть та же, но придётся работать с интегралами

Что мы видим в этой таблице, что внятно не проговаривается нигде:  
Первое. У нас есть всевозможные значения параметра θ. Этакой домен θ. Для каждого значения θ у нас есть вероятность p(θ). Т.е. у нас есть распределение вероятности! **Изначально** вероятность рассчитывается на основе распределения, которое мы задаем, например, внутри pm.Model (например, θ = pm.Beta или θ = pm.Normal). Но!!! Это, чтобы стартануть. Т.е. можно думать, что мы буквально для кажого значения θ рассчитвали и запихнули в таблицу и всё, про pm.Beta или pm.Normal забываем.  Теперь у нас есть просто данные с которыми работем. Или другими словами мы оторвались от аналитической записи и теперь плывем сами. Это будет важно для понимания последующего

Второе. У нас есть также домен и для X. Т.е. даже, если мы не встретим какой-то Х во входящих данных, то мы всё равно имеем для него расчёт. Опять же есть **изначальный** расчет. Получаем, этакое пространство θ,X. Какова вероятность p(X=?). Это определяется через pmf(X|θ). Т.е. тут условная вероятность. Иначе говоря: "Возьми конкретный х[i], возьми конкретный параметр θ[j] и рассчитай на основе модели значение (pmf/pdf)."

Третье. Последняя колонка в таблицу соответствует совместной вероятности p(θ) и pmf(x|θ). Это кстати объясняет, откуда берётся подынтегральное выражение в знаменателе справа. Тут в буквальном смысле просто прочтение таблицы.

Как теперь выглядит итеративное обновление апостериорного распределения. А вот так:

```
for current_x in X:  
  aposterior_distr = []  

  for theta in theta_range:  
    выбрать_из_таблицы_JPT_то_что_надо_для_расчёта_temp  
    temp = P(current_x|theta) * P(theta) / P(current_x)  
    aposterior_distr.append(temp)  

  #aposterior_distr = это есть новое распределение p(θ), которое уже "оторвано" от аналитической записи (живёт само по себе в виде данных)  
  обновить_таблицу_JPT, т.е. залить новые значения p(θ) в колонку p(θ) + пересчитать предпоследнюю и последнюю колонки
```


Другими словами. Мы рассматриваем каждый квант (экземпляр) входящих данные (X) через призму всех возможных значений θ. Сиречь примеряем различные варианты одной и той же модели к новому x. Таким образом получаем спектр того, насколько хорошо каждое из θ. Этому спектру соответствует какое-то распределение. Далее мы испольузуем это обонвленное распределения для новой порции данных и т.д. По идее это должно всё более сдвигать распределение в сторону истины.

Вот этот нюанс, что мы условно говоря заливаем новое распределение p(θ) в JPT, он долго не доходил до меня. Без перезаливки получалось бы, что мы вновь и вновь использовали то же распределение (например, stats.beta(a=200,b=1000) для каждой новой порции данных. И получалось бы топтание на месте. Вроде обработали x[i], получили новое распределение, а где оно используется для x[i+1] - непонятно и x[i+1] обрабатывалось бы с нуля (как будто не было никакой предыстории). А раз так, то становится понятно, почему изначально pm.DiscreteUniform может превратиться в какую нибудь пилу или домик - после изначального посева из аналитической формы, распределение в виде данных живёт само по себе и мутирует.

Про формулу Байеса и применимость её в PyMC

Первый рабочий прототип изучений трансформации бинарных сигналов. Очень много взято из "Логика сознания" Редозубова.

Первым делом подробнее разобрался с тем, как кодировать любой сигнал. И вырисовалась такая общая картина:
1) one-hot encoding (OHE). Каждое понятие кодируется одним битом. Соответственно понятий может быть максимум столько, сколько размерность вектора сигнала. Например, вектора размерности 256 - следовательно 256 понятий. Один-к-одному маппинг понятия и его визуализации (если мы говорим про картинки). Из способа кодирования вытекает, что один вектор сигнала может нести кол-во понятий равный его размерности. Т.е. 256 битный вектор может нести 256 понятий (всё биты выставлены в 1).

2) spared binary vectors (SBV). Понятие кодируется несколькими битами. Например, 6 битов из 256 или 9 битов из 2000. Здесь мы соприкасамемся с diadic/triadic memory. Такой способ кодирования даёт резкое увеличением кол-ва понятий. Например, 6 битов из 256 даёт возможность закодировать (256 * 255 * 254 * 253 * 252 * 251)  / 6! = 44 трлн. понятий. Вместе с ростом кол-ва понятий снижается кол-во понятий, которое можно запихать в вектор и после уверенно восстановить. См. картинку справа. Также здесь приходится вводить явный маппинг какому пикселю картинки соответствует та или иная комбинация битов.

3) hyper-dimensional vectors (HDV). Вариант SBV, когда размерность 10к и больше, а кол-во активных бит равна половине размерности. Высокая помехоустойчивость.
