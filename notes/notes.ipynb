Получается, что эмбеддинги или distributed representations - это кровь интеллектуальных систем. По-крайней мере трансформеров. Причины:
1) описывают сущность с многих сторон. И это не просто перечисление каких-то свойств. Это перечисление свойств, которые нужны для задач, которые система решает. Т.е. например, слово "red". Эмбеддинг будет передавать что слово передает "красность", что это прилагательное, что это анлийской слово, и т.д. Куча свойств или аспектов

2) distributed representation означает, что от одной точки (один разряд в OHE) мы переходим к этакому ёжику - много-много иколок, каждый из которых описывает тот или иной аспект сущности.

<img src="./img/2025-11-22/embeddings-1.jpg" width=800>

3) эмбеддинги позволяют описывать регулярность предметной области, описывать СТРУКТУРУ. Хорошо сказано в [ColahDLNLPRepr]:

   > "Intuitively, it feels a bit like the two languages have a similar ‘shape’ and that by forcing them to line up at different points, they overlap and other points get pulled into the right positions."

5) **обобщение базируется на эмбеддингах**!!! [ColahDLNLPRepr], [HintonLearnDistrRepr]

6) **out-of-vocabulary базируется на эмбеддингах**!!! [ColahDLNLPRepr], [BengioNeuralProbLangModel]

7) мутируя эмбеддинги за счёт простых алгебарических операций (сложение векторов, умножение на скаляр) можно получить новые осмысленные сущности

8) по факту трансформер - это про преобразование эмбеддингов. Эмебдиннги двигаются по пайплайну трансформера, и на каждом этапе эмбеддинг адаптируется/мутирует.  [TransCircMathFrmwrkForTransformer]

9) attention в трансформерах - это просто query по свойствам эмбеддингов: типа "есть какое-то прилагательное передо мной?". И тут становится очевидно, почему это работает - потому что каждый из токенов представлен эмбеддингом, которые соответствующие аспекты в себе описывает и который можно обычным cos_sim найти.

**Вывод**: формула мозга: суперкондишнинг (attention?) + эмбеддинги + символизм + абстрагирование

- backpropagation крайне маловероятен в мозгу
- фактографическая информация не хранится на синапсах. Да и вообще пластичность синапсов для обучения под вопросом
- вряд ли мозг умеет оперировать числами с плавающей запятов (embeddings) [OvrmnDTMem]

Что мы видим в этой таблице, что внятно не проговаривается нигде:  
Первое. У нас есть всевозможные значения параметра θ. Этакой домен θ. Для каждого значения θ у нас есть вероятность p(θ). Т.е. у нас есть распределение вероятности! **Изначально** вероятность рассчитывается на основе распределения, которое мы задаем, например, внутри pm.Model (например, θ = pm.Beta или θ = pm.Normal). Но!!! Это, чтобы стартануть. Т.е. можно думать, что мы буквально для кажого значения θ рассчитвали и запихнули в таблицу и всё, про pm.Beta или pm.Normal забываем.  Теперь у нас есть просто данные с которыми работем. Или другими словами мы оторвались от аналитической записи и теперь плывем сами. Это будет важно для понимания последующего

Второе. У нас есть также домен и для X. Т.е. даже, если мы не встретим какой-то Х во входящих данных, то мы всё равно имеем для него расчёт. Опять же есть **изначальный** расчет. Получаем, этакое пространство θ,X. Какова вероятность p(X=?). Это определяется через pmf(X|θ). Т.е. тут условная вероятность. Иначе говоря: "Возьми конкретный х[i], возьми конкретный параметр θ[j] и рассчитай на основе модели значение (pmf/pdf)."

Первым делом подробнее разобрался с тем, как кодировать любой сигнал. И вырисовалась такая общая картина:
1) one-hot encoding (OHE). Каждое понятие кодируется одним битом. Соответственно понятий может быть максимум столько, сколько размерность вектора сигнала. Например, вектора размерности 256 - следовательно 256 понятий. Один-к-одному маппинг понятия и его визуализации (если мы говорим про картинки). Из способа кодирования вытекает, что один вектор сигнала может нести кол-во понятий равный его размерности. Т.е. 256 битный вектор может нести 256 понятий (всё биты выставлены в 1).

2) spared binary vectors (SBV). Понятие кодируется несколькими битами. Например, 6 битов из 256 или 9 битов из 2000. Здесь мы соприкасамемся с diadic/triadic memory. Такой способ кодирования даёт резкое увеличением кол-ва понятий. Например, 6 битов из 256 даёт возможность закодировать (256 * 255 * 254 * 253 * 252 * 251)  / 6! = 44 трлн. понятий. Вместе с ростом кол-ва понятий снижается кол-во понятий, которое можно запихать в вектор и после уверенно восстановить. См. картинку справа. Также здесь приходится вводить явный маппинг какому пикселю картинки соответствует та или иная комбинация битов.

3) hyper-dimensional vectors (HDV). Вариант SBV, когда размерность 10к и больше, а кол-во активных бит равна половине размерности. Высокая помехоустойчивость.

Таким образом на одной части спектра лежит one-hot encoding, а на другой HDV. В задаче изучений трансформаций выяснилось, что кол-во гипотез, генерируемых на один входной сигнал напрямую зависит от кол-ва активных бит. Причём кол-во растёт если не факториально, то близко. Поэтому в память всё гипотезы не влезают очень скоро. HDV отметается. OHE в силу ограниченности набора понятий тоже отметается. Остаётся SBV. Надо было выбрать размерность и способ кодирования. Представьте себе пустую тёмную комнату. Есть два способа увидеть всё: у вас есть фонарик, и вы освещаете всё фонариком, но при этом, когда вы видите ботинки, вы не видите шляпные крючки и т.д. Есть второй способ-вы щёлкаете выключателем и видите всю комнату сразу. 

Этим и отличается умный от дурака. Дурак способен держать в сознании только одну мысль. Человек умный способен задействовать 4 канала информации. В диалектическом единстве, в полутонах, в противоречиях. 

Дурак не может понять, что такое единство и борьба противоречий. Дураку нужно конкретно - так надо за белых или за красных? Так ты за Россию или против? Так ты считаешь, что всё будет хорошо или плохо? 

Если ты говоришь, что каждое явление имеет свою плохую сторону и свою хорошую сторону и даже в такой страшной истории как война есть то хорошее, что человек видит кто чего стоит и люди делают самое большое на что они способны. Дурак считает, что Гитлер плохой, его надо было удушить в колыбельной. Означает ли это, что было плохо всё? Несмотря на это, он был храбр, он был блестящий оратор, он был необыкновенно харизматичен, он был прекрасный организатор и он был абсолютный патриот своей страны. Также с генералом Власовым. Он был предатель, изменник, повешен за дело. Но до этого он был прекрасным командиром. Умный скажет, что это так. А глупый скажет, что ты за Власова. 

Когда дурак видит что-то одно, всё остальное для него прекращает существовать

Генерализация частных случаев. Это когда мы на основе незначительных фактов делаем масштабные выводы. Заметили, что в офис привезли колотые печеньки, — делаем вывод, что компания уже не торт и разваливается.

Феномен Баадера — Майнхоф, или иллюзия частотности. Феномен заключается в том, что после возникновения какого-либо события, если мы встретим подобное ещё раз, оно воспринимается как необычайно частое. Например, вы купили синюю машину и с удивлением стали замечать, что вокруг очень много синих машин. Или пару раз увидели, что продакт-менеджеры ошиблись, — и впоследствии видите только то, что они ошибаются.

Предвзятость подтверждения, когда мы обращаем внимание только на ту информацию, которая подтверждает наши собственные взгляды, и не принимаем во внимание факты, этим взглядам противоречащие. Например, с негативными мыслями в голове обращаем внимание только на плохие события, а позитивные изменения в компании просто не замечаем.

Фундаментальная ошибка атрибуции: все гасконцы, а я д’Артаньян. Это когда ошибки других мы склонны объяснять их личными качествами, а достижения — везением, а в случае с собой — наоборот. Пример: коллега, положивший продакшен, — нехороший человек, а если положил я, значит, «не повезло, бывает».

Феномен справедливого мира, когда мы верим в то, что существует некая высшая справедливость, во имя которой все должны действовать.
