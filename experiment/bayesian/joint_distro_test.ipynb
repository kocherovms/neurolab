```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

# Бетабиномиальое распределение 

```python
n, a, b = (5, 100, 100)
x = np.arange(stats.betabinom.ppf(0, n, a, b) + 1, stats.betabinom.ppf(1, n, a, b) + 1)
px = stats.betabinom.pmf(x, n, a, b)
plt.bar(x, px)
plt.grid()
px.sum()
```

# Совм. распред. двух независимых биномиальных распред.

```python
n, px, py = (5, 0.5, 0.2)
x = np.arange(stats.binom.ppf(0, n, px) + 1, stats.binom.ppf(1, n, px) + 1)
y = np.arange(stats.binom.ppf(0, n, py) + 1, stats.binom.ppf(1, n, py) + 1)
px = stats.binom.pmf(x, n, px)
py = stats.binom.pmf(y, n, py)
plt.bar(x, px, label='px')
plt.bar(y, py, alpha=0.8, label='py')
plt.grid()
plt.legend()
px.sum(), py.sum()
```

```python
joint_p = px[:,np.newaxis] @ py[np.newaxis,:]
joint_p
```

```python
joint_p.sum()
```

```python
fig = plt.figure()
plt.figure(figsize=(12.5, 8))

X, Y = np.meshgrid(x, y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, joint_p, cmap='jet');
ax.view_init(azim=45);
```

```python
marginal_px = joint_p.sum(axis=1)
assert np.all(np.isclose(px, marginal_px))
px, marginal_px
```

```python
marginal_py = joint_p.sum(axis=0)
assert np.all(np.isclose(py, marginal_py))
py, marginal_py
```

# Совм. распред. независимых, но разных распред.

```python
n, px = (5, 0.2)
x = np.arange(stats.binom.ppf(0, n, px) + 1, stats.binom.ppf(1, n, px) + 1)
y = x.copy()
px = stats.binom.pmf(x, n, px)
py = np.array([0.25, 0, 0.05, 0.4, 0.2, 0.1])
plt.bar(x, px, label='px')
plt.bar(y, py, alpha=0.8, label='py')
plt.grid()
plt.legend()
px.sum(), py.sum()
```

```python
joint_p = px[:,np.newaxis] @ py[np.newaxis,:]
joint_p
```

```python
joint_p.sum()
```

```python
fig = plt.figure()
plt.figure(figsize=(12.5, 8))

X, Y = np.meshgrid(x, y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, joint_p, cmap='jet');
ax.view_init(azim=45);
```

```python
marginal_px = joint_p.sum(axis=1)
assert np.all(np.isclose(px, marginal_px))
px, marginal_px
```

```python
marginal_py = joint_p.sum(axis=0)
assert np.all(np.isclose(py, marginal_py))
py, marginal_py
```

```python
assert np.all(np.isclose(marginal_px[:,np.newaxis] @ marginal_py[np.newaxis,:], joint_p))
```

# Совм. распред. для примитивной Байесовской сети

```python
x = [1, 2, 3]
y = [1, 2, 3]
joint_p = np.array([[0, 0, 0.5], [0.3, 0, 0], [0.1, 0.1, 0]]).T

fig = plt.figure()
plt.figure(figsize=(12.5, 8))

X, Y = np.meshgrid(x, y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, joint_p, cmap='jet');
ax.view_init(azim=60);
```

```python
joint_p
```

```python
marginal_px = joint_p.sum(axis=0) # axis=0 because joint_p is transposed
marginal_py = joint_p.sum(axis=1)
marginal_px, marginal_py
```

```python
new_joint_p = marginal_px[:,np.newaxis] @ marginal_py[np.newaxis,:]
```

```python
fig = plt.figure()
plt.figure(figsize=(12.5, 8))

X, Y = np.meshgrid(x, y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, new_joint_p, cmap='jet');
ax.view_init(azim=60);
```

```python
new_marginal_px = new_joint_p.sum(axis=1) 
new_marginal_py = new_joint_p.sum(axis=0)
new_marginal_px, new_marginal_py
```
