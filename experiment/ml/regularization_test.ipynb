```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([1, 2, 3, 3])
h = np.array([1, 0.8, 2, 3])
slopes = np.linspace(-0.2, 1, 1000)
res_list = []
l1_pen_list = []
l2_pen_list = []
l1_err_list = []
l2_err_list = []
λ = 5

for slope in slopes:
    h_hat = 1 + slope * w
    res = (h - h_hat) ** 2
    l1_pen = λ * np.abs(slope) 
    l2_pen = λ * slope**2 
    res_list.append(np.sum(res))
    l1_pen_list.append(l1_pen * len(res))
    l2_pen_list.append(l2_pen * len(res))
    l1_err_list.append(np.sum(res + l1_pen))
    l2_err_list.append(np.sum(res + l2_pen))

res_list = np.array(res_list)
l1_pen_list = np.array(l1_pen_list)
l2_pen_list = np.array(l2_pen_list)
l1_err_list = np.array(l1_err_list)
l2_err_list = np.array(l2_err_list)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figwidth(16)
fig.set_figheight(6)

# ax1.set_title('L1 penalty')
ax1.plot(slopes, res_list, label='sq. residual')
ax1.plot(slopes, l1_pen_list, label='L1 penalty')
ax1.plot(slopes, l1_err_list, label='error')
ax1.set_xlabel('slope')
ax1.legend()

# ax2.set_title('L2 penalty')
ax2.plot(slopes, res_list, label='sq. residual')
ax2.plot(slopes, l2_pen_list, label='L2 penalty')
ax2.plot(slopes, l2_err_list, label='error')
ax2.set_xlabel('slope')
ax2.legend();
```
