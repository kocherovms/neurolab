```python
import cuml

# Install with desired log level before other imports
cuml.accel.install(log_level="debug")

from sklearn.linear_model import Ridge
from sklearn.svm import SVC, LinearSVC
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
```

# SVC Classification

```python
# Generate classification data
X_class, y_class = make_classification(n_samples=2000, n_features=20, n_informative=15, n_redundant=5, n_classes=3, random_state=42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

svc = SVC(kernel='rbf', C=1.0)
svc.fit(X_train_class, y_train_class)
svc_predictions = svc.predict(X_test_class)

svc = LinearSVC(C=1.0)
svc.fit(X_train_class, y_train_class)
svc_predictions = svc.predict(X_test_class)
```
